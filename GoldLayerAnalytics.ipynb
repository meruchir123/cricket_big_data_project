```python
from pyspark.sql.functions import *
```

```python
silver_df=spark.table('workspace.default.cricket_silver_current')
display(silver_df)
```

```python


#Match Type Distribution

gold_match_type_df=silver_df.groupBy('match_type').agg(count('*').alias("Total_Matches"))
display(gold_match_type_df)
```

```python
# Match count by stadium

gold_stadium_df=silver_df.groupBy('venue').agg(count('*').alias("Total_Matches_Stadium"))
display(gold_stadium_df)
```

```python
#Team Wise Match Count
from pyspark.sql.functions import col

gold_team1_df=silver_df.select(col("team_1").alias("team"))
gold_team2_df=silver_df.select(col("team_2").alias("team"))

all_teams_df=gold_team1_df.union(gold_team2_df)
gold_team_df=all_teams_df.groupBy('team').agg(count('*').alias("Total_Matches_by_Team"))
display(gold_team_df)
```

```python
%sql
select 
  count(*) as Total_matches, 
  count(distinct match_type) as Unique_Match_Types, 
  count(distinct venue) as Unique_Stadiums
from workspace.default.cricket_silver_current
```
